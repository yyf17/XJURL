## 2. 线性函数近似

```python
class LinearApprox:
    """线性函数近似"""
    def __init__(self, dim):
        self.weights = np.random.randn(dim) * 0.01
        self.bias = 0.0
    
    def forward(self, x):
        return np.dot(x, self.weights) + self.bias
    
    def update(self, x, target, lr=0.01):
        pred = self.forward(x)
        error = target - pred
        self.weights += lr * error * x
        self.bias += lr * error

def test_linear():
    app = LinearApprox(4)
    for _ in range(1000):
        x = np.random.randn(4)
        true_y = np.dot(x, [1,2,3,4]) + 0.5
        app.update(x, true_y, lr=0.01)
    print(f"权重: {app.weights}")
    print(f"偏置: {app.bias:.3f}")
    return app

test_linear()
```

## 3. 神经网络近似

```python
class NeuralNetApprox:
    """简单神经网络"""
    def __init__(self, input_dim, hidden_dim, output_dim):
        self.fc1 = np.random.randn(input_dim, hidden_dim) * 0.1
        self.fc2 = np.random.randn(hidden_dim, output_dim) * 0.1
    
    def forward(self, x):
        x = np.tanh(np.dot(x, self.fc1))
        x = np.dot(x, self.fc2)
        return x
    
    def relu(self, x):
        return np.maximum(0, x)

print("神经网络已定义")
```

## 5. 编程题

### 编程题1：线性TD学习

```python
class LinearTDAgent:
    """线性TD学习"""
    def __init__(self, state_dim, alpha=0.1, gamma=0.9):
        self.weights = np.zeros(state_dim)
        self.alpha = alpha
        self.gamma = gamma
    
    def predict(self, state):
        return np.dot(state, self.weights)
    
    def update(self, state, reward, next_state):
        td_error = reward + self.gamma * self.predict(next_state) - self.predict(state)
        self.weights += self.alpha * td_error * state
        return td_error

def test_linear_td():
    agent = LinearTDAgent(4)
    errors = []
    for _ in range(1000):
        s = np.random.randn(4)
        ns = np.random.randn(4)
        r = np.random.randn()
        err = agent.update(s, r, ns)
        errors.append(abs(err))
    print(f"平均TD误差: {np.mean(errors):.4f}")
    return errors

test_linear_td()
```
